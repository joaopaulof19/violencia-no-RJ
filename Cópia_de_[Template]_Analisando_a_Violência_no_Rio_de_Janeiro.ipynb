Custou acreditar em alguns n√∫meros estat√≠sticos obtido desse DataFrame. 

- Acontece em m√©dia 500 homicidios doloso por m√™s. üò±
- M√©dia de 71 homicidios por interven√ß√£o policial por m√™s. üò∞
- M√©dia de 318 tentativas de homicidios por m√™s. Isso significa que a cada 2 horas, h√° uma tentativa de algu√©m tirar a vida de outra pessoa. üò®
- M√©dia de 730 estupros por m√™s. Um estupro a cada duas horas. üòß

- 66 cadaveres encontrados por m√™s. Cerca de 2 ccorpos por dia. üíÄ

N√∫meros assustadores.

```
roubo_veiculo = df.roubo_veiculo.mean()
furto_veiculos = df.furto_veiculos.mean()
recuperacao_veiculos = df.recuperacao_veiculos.mean()

print(f'M√©dia de roubos de ve√≠culos: {roubo_veiculo}')
print(f'M√©dia de furto de ve√≠culos: {furto_veiculos}')
print(f'M√©dia de recupera√ß√£o de ve√≠culos: {recuperacao_veiculos}\n')

carros_rec_perc = recuperacao_veiculos/(roubo_veiculo + furto_veiculos)
print(f'porcentagem de carros recuperados em rela√ß√£o aos carros roubados + carros furtados: {carros_rec_perc * 100} %\n')

# Encontre os valores m√°ximos (max) e m√≠nimos (min) da coluna hom_doloso
print(f'N√∫mero de homicidios dolosos do m√™s mais tranquilo: {df.hom_doloso.min()}')
print(f'N√∫mero de homicidios dolosos do m√™s mais conturbado {df.hom_doloso.max()}')
```
